```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from plotly.offline import iplot
import plotly.graph_objs as go

from google.colab import drive
drive.mount('/content/drive')
```

```python
timesData = pd.read_csv("/content/timesData.csv")
```

```python
timesData.head(9)
```

```python
df2013 = timesData[timesData.year == 2013].iloc[:5,:]
df2013
```

```python
trace1 = go.Bar(x = df2013.university_name,
                y = df2013.teaching,
                name = "teaching",
                marker = dict(color = 'rgba(255,121,149,0.4)',
                line=dict(color='rgb(0,0,0)',width=2.3)),text = df2013.country)
```

```python
trace2 = go.Bar(x = df2013.university_name,
                y = df2013.international_students,
                name = "international_students",
                marker = dict(color = 'rgba(255,191,109,0.6)',
                line=dict(color='rgb(0,0,0)',width=2.3)),text = df2013.country)
```

```python
data = [trace1, trace2]
layout = go.Layout(barmode = "group")
fig = go.Figure(data = data, layout = layout)
iplot(fig)

```
